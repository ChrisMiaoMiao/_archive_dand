# 清除示例 II
我们先练习癌症数据缺失值和重复值的处理。

```python
import pandas as pd
# 读入 `cancer_data.csv`
df = pd.read_csv('cancer_data.csv')

# 用 info() 检查哪些列有缺失值
df.info()
```

```python
# 用均值填充缺失值
texture_mean = df['texture_mean'].mean()
df['texture_mean'].fillna(texture_mean, inplace=True)
smoothness_mean = df['smoothness_mean'].mean()
df['smoothness_mean'].fillna(smoothness_mean, inplace=True)
symmetry_mean = df['symmetry_mean'].mean()
df['symmetry_mean'].fillna(symmetry_mean, inplace=True)
texture_SE_mean = df['texture_SE'].mean()
df['texture_SE'].fillna(texture_SE_mean, inplace=True)
smoothness_SE_mean = df['smoothness_SE'].mean()
df['smoothness_SE'].fillna(smoothness_SE_mean, inplace=True)
symmetry_SE_mean = df['symmetry_SE'].mean()
df['symmetry_SE'].fillna(symmetry_SE_mean, inplace=True)
texture_max_mean = df['texture_max'].mean
df['texture_max'].fillna(texture_max_mean, inplace=True)
smoothness_max_mean = df['smoothness_max'].mean()
df['smoothness_max'].fillna(smoothness_max_mean, inplace=True)
symmetry_max_mean = df['symmetry_max'].mean()
df['symmetry_max'].fillna(symmetry_max_mean, inplace=True)



# 用 info() 确认修改
df.info()
```

```python
# 检查数据中的重复
sum(df.duplicated(subset='id'))
```

```python
# 丢弃重复
df.drop_duplicates(subset='id', inplace=True)
```

```python
# 再次检查数据中的重复，确认修改
sum(df.duplicated(subset='id'))
```

## 重命名列
由于之前修改了数据集，使其仅包括肿瘤特征的均值，因此每个特征末尾好像不需要 "_mean" 。而且，稍后输入分析还要多耗费时间。我们现在想一些要分配给列的新标签。

```python
# 从列名称中移除 "_mean"
new_labels = []
for col in df.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])  # 不包括最后 6 个字符
    else:
        new_labels.append(col)

# 列的新标签
print(new_labels)
```

```python
# 为数据框中的列分配新标签
df.columns = new_labels
print(df.columns)

# 显示数据框的前几行，确认更改
df.head()
```

```python
# 将其保存，供稍后使用
df.to_csv('cancer_data_edited.csv', index=False)
```
