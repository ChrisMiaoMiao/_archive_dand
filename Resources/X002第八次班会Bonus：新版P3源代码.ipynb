### 2.3 数据表中的空值清洗

第二个要处理的问题是数据表中的空值，在python中空值被显示为NaN。在处理空值之前我们先来检查下数据表中的空值数量。对于一个小的数据表，我们可以人工查找，但对于较为庞大的数据表，就需要寻找一个更为方便快捷的方法了。  
Pandas中查找数据表中空值的函数有两个，一个是函数isnull，如果是空值就显示True。另一个函数notnull正好相反，如果是空值就显示False。  
此处采用isnull的方法循环：

```python
column_list = [column for column in df]

# Inform use of non-numm type exclusion
print('P.S. Exclude non-null column\n')
# Extract null values through iterations
column_to_purse = []
for column in column_list:
    if len(df.isnull()[column].value_counts().index) == 2:
        print('The null values of column {} is:'.format(column))
        print(df.isnull()[column].value_counts())
        column_to_purse.append(column)
        print('\n')
```

可以看到，存在null值的column分别为：imdb_id，cast，homepage，director，tagline，keywords，overview，genres和production_companies。  
由于一些column无法采用填入0的方法进行处理（如URL），因此对这些数据中的null值进行清除：

```python
# Purse null values
df = df.dropna()
# Exam zero value again
for number in df.isnull().sum():
    if number != 0:
        print("The parsing is not completed yet.")
```

去除后的数据已经不含有空值了。

### 2.4 数据中的字符串拆分

由于数据中cast等项中的每一条都包含许多条目，以“|”作为分割，所以需要进行格式的转写，方便后面的查询和探索性数据分析：

```python
# Splitting cast column
for column in column_list:
    try:
        if '|' in df[column][0]:
            df[column] = df[column].apply(lambda x:x.split('|'))
    except:
        pass
# Print result
df.head()
```

可以看到，此时的cast已经被划分成了list的形式，方便后面的探索了。

### 2.6 数据中的异常值和极端值

第五个要处理的问题是数据中的异常值和极端值，发现异常值和极端值的方法是对数据进行描述性统计。使用describe函数可以生成描述统计结果。其中我们主要关注最大值(max)和最小值(min)情况

```python
# Describe the dataframe to see abnormal columns
df.describe()
```

可以看到，其中存在许多异常点问题，比如budget、runtime中存在0时长，release_year过小等问题，以下逐一处理：

首先去除所有的零值：

```python
# Find all zero values
column_mins = df.describe().min()
# storage zero existed columns
column_zero_existed = []
column_index = np.where(column_mins == 0)
for index in column_index:
    column_zero_existed.append(column_mins.index[index])
# Convert this array
column_zero_existed = np.array(column_zero_existed[0])
# Parse zero values
for column in column_zero_existed:
    df = df[df[column]> 0]
```

```python
# Describe the dataframe to see abnormal columns
df.describe()
```

可以看到，零值存在于budget，runtime，budget_adj，revenue_adj中，并得到了清除。  
在进行EDA之前重复一遍NaN值的清除：

```python
# Purse null values
df = df.dropna()
# Exam zero value again
for number in df.isnull().sum():
    if number != 0:
        print("The parsing is not completed yet.")
```

### 3.1 单变量探索

* 电影的收入分布

对于电影行业来说，收入一直是外界猜测的话题，通过IMDB数据集，我们可以轻松得到电影收入的分布情况。将收入数据进行六档分级，可以更好的看到收入的规律。根据五类描述法，将收入分成六个部分并画图：

```python
# Ignore warning
import warnings
warnings.filterwarnings("ignore")

# Get soft encoding from quantiles
revenue_encodings = []
encoding_names = ['Mountain Top', 'High', 'Medium', 'Below Average', 'Low']
Q_top = df['revenue'].quantile(1)
for i in range(6):
    revenue_encodings.append(Q_top/5*i)

# Drop extraneous data
revenue_encodings = np.sort(np.array(revenue_encodings))

# Plot the histogram of movies revenue
df['revenue_ranking'] = pd.cut(df['revenue'], bins=revenue_encodings, labels=encoding_names)

# Plot a histogram to present the pattern
revenue_ranking = Counter(df['revenue_ranking'])
revenue_ranking = dict(revenue_ranking)
for i, key in enumerate(revenue_ranking):
    plt.bar(i, revenue_ranking[key], color='blue')

# Add title and labels
plt.title('Revenue ranking of movies')
plt.xlabel('Revenue categories')
plt.ylabel('Frequency')
```

可以看出，绝大部分电影的收入都在行业内最赚钱电影的20%以下，可见整个行业的收入相对而言是赢者通吃的节奏。

* 演员主演的影片数量分布

众所周知，演员界有许多劳模级的人物，无数次陪跑终于拿到奥斯卡的小李子就是其中之一。出于好奇，想要对演员在整个演艺生涯中主演的影片数量进行统计和分析。

利用数据清洗过程中得到的演员数据，可以进行探索性数据分析：

```python
# Utilization of cast data
cast_list = []
for casts in df['cast']:
    for cast in casts:
        cast_list.append(cast)

# Collect cast number via Counter
cast_counter = Counter(cast_list)
cast_number = []
for number in cast_counter.values():
    cast_number.append(number)
cast_number = Counter(cast_number)
```

画出演员出演频率图：

```python
# Plot frequency of casts
cast_number = dict(cast_number)
for i, key in enumerate(cast_number):
    plt.bar(i, cast_number[key], color='blue')
plt.title('Frequency of casting')
plt.xlabel('Frequency')
plt.ylabel('Number of casting')
```

可以看到，大部分演员在演艺生涯中会出演四部电影，而小李子大概演过二十多部，在这个图里面应该是特别靠右的存在，因此拿到奥斯卡可以说只是时间问题。
